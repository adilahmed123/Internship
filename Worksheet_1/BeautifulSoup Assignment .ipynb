```python
#programme to display IMDB's top 100 movies data
```

```python
import pandas as pd
```

```python
url = "https://www.imdb.com/list/ls091520106/"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
movie_name = []
year = []
rating = []
```

```python
movie = soup.find_all('div',class_ = "lister-item-content")
for store in movie:
    movies = store.h3.a.text
    movie_name.append(movies)
    
    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    year.append(year_of_release)
    
    Rating = store.find('span',class_ = 'ipl-rating-star__rating').text.replace('\n','')
    rating.append(Rating)
    
```

```python
df = pd.DataFrame({ 'Name' : movie_name , 'Year_of_release' : year , 'Rating' : rating})
df
```

```python
# 4) python programme to scrape the Womens cricket team data
# a) Top 10 ODI teams of womens 
```

```python
url = 'https://www.icc-cricket.com/rankings/womens/team-rankings/odi'
page = requests.get(url)
```

```python
bs1 = BeautifulSoup(page.content,'html.parser')
```

```python
wo = []
w = bs1.find_all('span',class_ = 'u-hide-phablet')
for data in w:
    w1 = data.text
    wo.append(w1)
wo[0:10]
```

```python
matches = []
points = []
rating = []
```

```python
for i in bs1.find_all('td', class_ = 'table-body__cell u-center-text'):
    rating.append(i.text)
for i in range(0,len(rating)-1,2):
    matches.append(rating[i])
    points.append(rating[i+1])
```

```python
a = {'Womens_teams' : wo[0:10] , 'Matches' : matches , 'Points' : points , 'Rating' : rating[1:2:2]}
df = pd.DataFrame.from_dict(a,orient = 'index')
df = df.transpose()
df
```

```python
# programme to scrape the details from dineout.co.in
```

```python
url = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'
page = requests.get(url)
```

```python
bs6 = BeautifulSoup(page.content,'html.parser')
```

```python
Resturant = []
r = bs6.find_all('a' , class_ = 'restnt-name ellipsis')
for data in r:
    res = data.text
    Resturant.append(res)
Resturant
```

```python
cuisine = []
c = bs6.find_all('span', class_ = 'double-line-ellipsis')
for data in c:
    cui = data.text.split('|')[1]
    cuisine.append(cui)
cuisine
```

```python
Location1 = []
l = bs6.find_all('div', class_ = 'restnt-loc ellipsis')
for data in l:
    lo = data.text
    Location1.append(lo)
Location1
```

```python
Ratings = []
r1 = bs6.find_all('div', class_ = 'restnt-rating rating-4')
for data in r1:
    rat = data.text
    Ratings.append(rat)
Ratings
```

```python
images = []
for data in bs6.find_all('img' , class_ = 'no-img'):
    images.append(data['data-src'])
images
```

```python
df = pd.DataFrame({'Resturant_Name' : Resturant , 'Cuisine' : cuisine , 'Location' : Location1 , 'Rating' : Ratings, 'Images' : images})
df
```
