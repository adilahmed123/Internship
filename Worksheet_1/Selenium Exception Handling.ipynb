```python
# 1) python programme to scrape guitars details from Amazon website including
# a) Brand Name b) Name of the product c) Price d) Return/Exchange e) Expected Delivery f) Avaialblity g) product URL
```

```python
driver = webdriver.Chrome(r"C:\\Users\\adil2\\Downloads\\chromedriver_win32\\chromedriver.exe")
```

```python
driver.get('https://www.amazon.in/')
```

```python
Brand_Name = []
Product_Name = []
Price = []
Replacement = []
Exp_delivery =[]
Availablity = []
url = []
```

```python
link = "https://www.amazon.in/s?k=guitars&crid=2QKN6SGC3RB8N&sprefix=guitars%2Caps%2C585&ref=nb_sb_noss_1"
driver.get(link)
```

```python
for data in driver.find_elements_by_xpath('//a[@class = "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]'):
    url.append(data.get_attribute('href'))
url
```

```python
for data in url:
    driver.get(data)
    try:
        brand = driver.find_element_by_xpath('//td[@class = "a-span9"]')
        Brand_Name.append(brand.text)
    except:
        Brand_Name.append('-')
Brand_Name
```

```python
for data in url:
    driver.get(data)
    try:
        prd_na = driver.find_element_by_xpath('//span[@class = "a-size-large product-title-word-break"]')
        Product_Name.append(prd_na.text)
    except:
        Product_Name.append('-')
Product_Name
```

```python
for data in url:
    driver.get(data)
    try:
        pr = driver.find_element_by_xpath('//span[@class = "a-price-whole"]')
        Price.append(pr.text)
    except:
        Price.append('-')
Price
```

```python
for data in url:
    driver.get(data)
    try:
        rep = driver.find_element_by_xpath('//a[@class = "a-size-small a-link-normal a-text-normal"]')
        Replacement.append(rep.text)
    except:
        Replacement.append('-')
Replacement
```

```python
for data in url:
    driver.get(data)
    try:
        exp = driver.find_element_by_xpath('//div[@id = "ddmDeliveryMessage"]')
        Exp_delivery.append(exp.text)
    except:
        Exp_delivery.append('-')
Exp_delivery
```

```python
for data in url:
    driver.get(data)
    try:
        avail = driver.find_element_by_xpath('//span[@class = "a-size-medium a-color-success"]')
        Availablity.append(avail.text)
    except:
        Availablity.append('-')
Availablity
```

```python
ama = {'Brand_Name': Brand_Name,'Product_Name':Product_Name,'Price':Price,'Replacement/Return':Replacement,'Expected_delivery':Exp_delivery,'Availablity':Availablity,'url':url}
df = pd.DataFrame.from_dict(ama,orient = 'index')
df = df.transpose()
df.head()
```

```python
# 10) Programme to scrape a data of all avialble hostels from www.hostelsworld.com
```

```python
driver = webdriver.Chrome(r"C:\\Users\\adil2\\Downloads\\chromedriver_win32\\chromedriver.exe")
```

```python
driver.get('https://www.hostelworld.com/')
```

```python
url = 'https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-01-04&to=2022-01-07&guests=2&page=1'
driver.get(url)
```

```python
name = []
na = driver.find_elements_by_xpath('//h2[@class = "title title-6"]')
for data in na:
    nam = data.text
    name.append(nam)
name
```

```python
distance = []
dis = driver.find_elements_by_xpath('//span[@class = "description"]')
for data in dis:
    dist = data.text.split('-')[1]
    distance.append(dist)
distance
```

```python
rating =[]
ra = driver.find_elements_by_xpath('//div[@class = "score orange big"]')
for data in ra:
    rat = data.text
    rating.append(rat)
rating
```

```python
total_review = []
tr = driver.find_elements_by_xpath('//div[@class = "reviews"]')
for data in tr:
    trv = data.text
    total_review.append(trv)
total_review
```

```python
overall_review = []
ovr = driver.find_elements_by_xpath('//div[@class = "keyword"]')
for data in ovr:
    orv = data.text
    overall_review.append(orv)
overall_review
```

```python
Private_price = []
pp = driver.find_elements_by_xpath('//div[@class = "price title-5"]')
for data in pp:
    priv = data.text
    Private_price.append(priv)
Private_price
```

```python
facilty = []
fa = driver.find_elements_by_xpath('//span[@class = "label"]')
for data in fa:
    faci = data.text
    facilty.append(faci)
facilty
```

```python
lon = {'Hostel_Name':name,'Distance_from_city_centre':distance,'Hostel_Rating':rating,'Total_Review':total_review,'Overall_Review':overall_review,'price':Private_price,'Facilities':facilty}
df = pd.DataFrame.from_dict(lon,orient = 'index')
df = df.transpose()
df.head()
```
